After downloading the input dataset and viewing their directory output, I copied these respective directories to create DataFrame Objects, ***train_data*** and ***test_data*** in order to view my train.csv and test.csv files. 

For my training Datasets, I noticed that the 'Age' column contains lots of missing data. Therefore, I decided to create a Pandas boolean object, ***bool_series*** 

The **bool_series** object will allow me to access only the passengers of my training Datasets with *complete* details. By doing this, I will train my Machine learning model with a complete Dataset.

```python
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
bool_series = pd.notnull(train_data['Age'])
train_data = train_data[bool_series]
train_data
```

```python
features = ['Pclass', 'Male', 'Age', 'SibSp', 'Parch', 'Fare']
train_data['Male'] = train_data['Sex'] == 'male'


y = train_data['Survived'].values
X = train_data[features].values

print(X)
```

A Machine Learning model for the Titanic DataSet will be built using Random Forest.

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)
rf.fit(X,y)
rf.score(X,y)
```

For my test Dataset, there are still missing data values in the 'Age column. But I cannot ignore rows that contain missing 'Age' values because it will not allow me to make predictions for such rows.

Therefore, I will fill in those missing data in the 'Age' column by using the **fillna()** function. This function will fill up missing data in a Dataframe with value of the function's argument as shown below:

```python
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')

features = ['Pclass', 'Male', 'Age', 'SibSp', 'Parch', 'Fare']
test_data['Male'] = test_data['Sex'] == 'male'

test_data = test_data.fillna(0)
X_test = test_data[features].values
test_data
```

The missing data in the Test Dataset was filled with '0'. After this,the prediction of survival for the 418 passengers in the test Dataset is stored and saved as submission_titanic.csv:

```python
y_pred = rf.predict(X_test)
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': y_pred})
output.to_csv('submission_titanic.csv', index=False)
print("Submission file created successfully")
```
